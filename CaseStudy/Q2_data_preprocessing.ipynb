```
from pyspark.sql import SparkSession, functions as F
```

```
spark = SparkSession.builder.appName('master').getOrCreate()
```

```
gender = spark.read.csv('/FileStore/tables/Gender_StatsData.csv', inferSchema=True, header=True).drop('_c66')
gender.show(2)
```

```
gender.count()
```

```
years = gender.columns[4:]
```

```
len(years)
```

```
year_str = f"{len(years)}, " + ', '.join([f"'{year}', `{year}`" for year in years])
```

```
f"stack({year_str}) as (Year,Value)"
```

```
# Use stack function to transform column names(years) into a colummn - Year and their corresponding values to the column Value
gender_new = gender.select("Country Name", "Country Code", "Indicator Name", "Indicator Code",
F.expr(f"stack({year_str}) as (Year,Value)")).where("Value is not null")

# Write the preprocessed data to csv file
# gender_new..coalesce(1).write.csv('FileStore/tables/gender_unpivot.csv')
```
